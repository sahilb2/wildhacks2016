```python

from clarifai.rest import ClarifaiApp
app = ClarifaiApp()
exlink = "http://i.imgur.com/YZDcqH9.jpg"

model = app.models.get('NSFWBlock')
model.train()
def checkNSFWLink(link):
    nsfw_data =  model.predict_by_url(link)['outputs'][0]['data']['concepts']
    if nsfw_data[0]['name'] == 'safe':
        return "SFW"
    else:
        return "NSFW"
print checkNSFWLink(exlink)

```
